# Cicli

```python
numbers_new = [34.5, 65.3, 78.2, 32, 12, 14]
for elemeto in numbers_new:
  print(f"elemento originale {elemeto}")
  print(f"elemento incrementato {elemeto+50}")
```

```python
for i in range(8):
  print(f'ciao {i}')
```

```python
for i in range(3,8):
  print(f'ciao {i}')
```

```python
numbers_new = [34.5, 65.3, 78.2, 32, 12, 14]
for elemeto in numbers_new:
  print(f"elemento originale {elemeto}")
  print(f"elemento incrementato {elemeto+50}")
  for i in range(3,8,2):
    print(f'ciao {i}')
```

```python
var = 0
for i in range(8):
  var = var + 0.5
  print(f'ciao {var}')
```
