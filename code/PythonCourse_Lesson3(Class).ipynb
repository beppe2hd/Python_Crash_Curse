```python
class Cat():
  def __init__(self, assigned_name, assigned_age):
    self.name = assigned_name
    self.age = assigned_age

  def say_name(self):
    print(f"My name is {self.name}")

  def change_name(self, new_name):
    self.name = new_name

  def birtday(self):
    self.age += 1

  def say_age(self):
    print(f"I'm {self.age} years old")

  def say_name_many_times(self, rep):
    for i in range(rep):
      print(f"My name is {self.name}")

  def say_name_many_times_2(self, rep):
    for i in range(rep):
      self.say_name()
      self.say_age()
```

```python
creatura4 = Cat("Spillo", 2)
```

```python
print(creatura4.name)
creatura4.change_name("Rocco")
print(creatura4.name)
```

Correntista

```python
class Correntista():

  def __init__ (self, nome, cognome, saldo_disp, saldo_inv):
    self.nome = nome
    self.cognome = cognome
    self.saldo_disp = saldo_disp
    self.saldo_inv = saldo_inv

  def preleva(self, prelievo):
    nuovo_saldo_disp = self.saldo_disp - prelievo
    self.saldo_disp = nuovo_saldo_disp
    print(f'Il nuovo saldo disponibile è {self.saldo_disp}')

  def investi(self, investimento):
    if self.saldo_disp > investimento:
      self.saldo_disp = self.saldo_disp - investimento
      self.saldo_inv = self.saldo_inv + investimento
      print(f'Il saldo investito è {self.saldo_inv}')
    else:
      print('Non hai abbastanza credito')
```

```python
corr1 = Correntista( 'Marco', 'Del Coco', 500, 300)
```

```python
corr1.nome
```

```python
corr1.saldo_disp
```

```python
corr1.preleva(100)
```

```python
corr1.saldo_disp
```

```python
corr1.investi(2000)
```

```python
corr1.saldo_disp
```

```python
corr1.investi(200)
```

```python
class Correntista():

  def __init__ (self, nome, cognome, saldo_disp, saldo_inv):
    self.nome = nome
    self.cognome = cognome
    self.saldo_disp = saldo_disp
    self.saldo_inv = saldo_inv

  def _aggiorna_saldo(self, val_agg):
    nuovo_saldo_disp = self.saldo_disp - val_agg
    self.saldo_disp = nuovo_saldo_disp

  def preleva(self, prelievo):
    self._aggiorna_saldo(prelievo)
    print(f'Il nuovo saldo disponibile è {self.saldo_disp}')

  def investi(self, investimento):
    if self.saldo_disp > investimento:
      self._aggiorna_saldo(investimento)
      self.saldo_inv = self.saldo_inv + investimento
      print(f'Il saldo investito è {self.saldo_inv}')
    else:
      print('Non hai abbastanza credito')
```

```python
corr2 = Correntista("M", "DC", 400, 150)
```

```python
corr2.preleva(10)
```

```python
corr2._aggiorna_saldo(30)
```

```python
corr2.saldo_disp
```

```python
class CorrentistaCrypto(Correntista):
  def investi_in_cripto(self, valore):
    self.saldo_disp = self.saldo_disp - valore
    print(f"Hai inventito {valore} in cryoto valute")
```

```python
corr3 = CorrentistaCrypto("m", 'c', 45, 67)
```

```python
corr3.saldo_disp
```

```python
corr3.investi_in_cripto(3)
```
