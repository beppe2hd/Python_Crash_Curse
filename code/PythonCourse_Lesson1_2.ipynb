## Funzioni

```python
price1 = 100
discount1 = 10

final_price1 = price1 - (price1 * discount1 / 100)
print(f"Product 1 final price: {final_price1}")

price2 = 250
discount2 = 10

final_price2 = price2 - (price2 * discount2 / 100)
print(f"Product 2 final price: {final_price2}")

price3 = 80
discount3 = 10

final_price3 = price3 - (price3 * discount3 / 100)
print(f"Product 3 final price: {final_price3}")
```

```python
def calcola_sconto(price, discount):
  final_price = price - (price * discount / 100)
  print(f"Product final price: {final_price}")
```

```python
calcola_sconto(price1, discount1)
calcola_sconto(price2, discount2)
calcola_sconto(price3, discount3)
calcola_sconto(540, 22)
```

```python
def calcola_sconto(price, discount):
  final_price = price - (price * discount / 100)
  return final_price
```

```python
prezzo = calcola_sconto(100, 35)
print(f"Product final price: {prezzo}")
```

```python
def calcola_sconto(price, discount):
  final_price = price - (price * discount / 100)
  saved = (price * discount / 100)
  return final_price, saved
```

```python
prezzo, risparmio = calcola_sconto(100, 35)
print(prezzo)
print(risparmio)
```
